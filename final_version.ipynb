## Preprocessing

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Components Analysis

At this point, we¬¥ll use two Component Analysis techniques, PCA and t-SNE. Principal Component Analysis, or PCA, is a dimensionality-reduction method that is often used to reduce the dimensionality of large data sets, by transforming a large set of variables into a smaller one that still contains most of the information in the large set. The secont technique, t-SNE, t-Distributed Stochastic Neighbor Embedding, is a technique for dimensionality reduction that is particularly well suited for the visualization of high-dimensional datasets. The technique can be implemented via Barnes-Hut approximations, allowing it to be applied on large real-world datasets. 

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_train)

tsne = TSNE(n_components=3)
X_tsne = tsne.fit_transform(X_train)
```

```python
plt.figure(figsize=(18, 9), dpi=80)
plt.subplot(1,2,1)
plt.title('PCA')
for y in np.unique(y_train):
    X_tmp = X_pca[y_train == y]
    plt.scatter(X_tmp[:, 0], X_tmp[:, 1], label=y, alpha=0.75)
plt.legend()
plt.subplot(1,2,2)          
plt.title('TSNE')
for y in np.unique(y_train):
    X_tmp = X_tsne[y_train == y]
    plt.scatter(X_tmp[:, 0], X_tmp[:, 1], label=y, alpha=0.75)
plt.legend()

plt.show()
```

It is relevant to point that when appliyng PCA, the distinction of the three classes are noticiable, in which its separation is almost perfect. By using TSE tecnique, in the other hand, the separation is not well defined, when considered only the 2 major components.

## Baseline and Batch

```python
clf = LogisticRegression()
clf.fit(X_train, y_train)
```

```python
y_test_pred = clf.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
plot_confusion_matrix(clf, X_test, y_test, normalize='true')
```

```python
losses = clf.losses()

# visualising losses
plt.figure(figsize=(18, 9), dpi=80)
plt.plot(losses)
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()
```

By using base model, the performance does not look good. The model was trained considering standard values for LogisticRegression algorithm. On the next section, a fine-tune will be applied aiming to obtain better results.

## Stochastic

```python
clf = LogisticRegression(optimizer="stochastic")
clf.fit(X_train, y_train)
```

```python
y_test_pred = clf.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
plot_confusion_matrix(clf, X_test, y_test, normalize='true')
```

```python
losses = clf.losses()

# visualising losses
plt.figure(figsize=(18, 9), dpi=80)
plt.plot(losses)
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()
```

## Mini-batch

```python
clf = LogisticRegression(optimizer="mini")
clf.fit(X_train, y_train)
```

```python
y_test_pred = clf.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
plot_confusion_matrix(clf, X_test, y_test, normalize='true')
```

```python
losses = clf.losses()
plt.figure(figsize=(18, 9), dpi=80)

# visualising losses
plt.plot(losses)
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()
```

By comparing the three approaches, it is clear that all three converges. What differs is the time to converse for each approach. The stochastic model will result in a much higher number of iteractions given its nature. In the other hand it seems at the end of the execution results on lesser error than batch approach. By looking into mini-batch, it seems to aggregate the best of the other two, it results on the lower error at the end of execution and given the fact that algorithm implements a batch size of 20.


## Fine-Tuning

```python
n_jobs = 4
n_iter = 100
cv = 5
scoring = 'balanced_accuracy'
```

```python
# LogisticRegression:
param_grid = {'learning_rate': [10, 1, 1e-1, 1e-2, 1e-3],
              'n_epochs': [100, 200, 500, 1000],
              'alpha': [10, 1, 1e-1, 1e-2, 1e-3],
             }

logreg_clf = RandomizedSearchCV(LogisticRegression(), param_distributions=param_grid, n_iter=n_iter, cv=cv,
                             scoring=scoring, n_jobs=n_jobs, verbose=1, random_state=42)
logreg_clf.fit(X_train, y_train)
```

```python
model = logreg_clf.best_estimator_ 
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test)
print(classification_report(y_test, y_test_pred))
```

```python
plot_confusion_matrix(model, X_test, y_test, normalize='true')
```

```python
losses = model.losses()
plt.figure(figsize=(18, 9), dpi=80)

# visualising losses
plt.plot(losses)
plt.xlabel("Number of iterations")
plt.ylabel("Loss")
plt.show()
```

It is clear that the fine-tune step is essential to acchieve a better final result. Comparing the results using fine-tune by results previously obtainated, the class prediction was flawless and checking the lossess, it it clear that algorithm converged after less than 10 iteractions. 
Of course, as this is a very simple dataset, there¬¥s no guarantee that same results will be seen on a complex one.
